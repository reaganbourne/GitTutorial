```python
start = 0 
pgnum = 0
for idx, cell in enumerate(tot_json['cells']):

    print(cell['source'])
    if '<!-- section -->' in cell['source']:
        print(idx)
        j = open("ok.ipynb")
        new_json = json.loads(j.read())
        new_json['cells'] = tot_json['cells'][start:idx+1]
        pgnum += 1
        with open(f"Git_Tutorial{pgnum}.ipynb", "w") as outfile:
            json.dump(new_json, outfile)
            print(start)
            print(idx)
            print(pgnum)
            print("\n\n\n")
        start = idx
        print(str(idx) + "\n\n")

    elif idx == len(tot_json['cells']) -1:
        j = open("ok.ipynb")
        new_json = json.loads(j.read())
        new_json['cells'] = tot_json['cells'][start:idx]
        pgnum += 1
        with open(f"Git_Tutorial{pgnum}.ipynb", "w") as outfile:
            json.dump(new_json, outfile)
    
        
    
```
